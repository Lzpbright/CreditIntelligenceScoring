```python
import pandas as pd
import numpy as np

data = pd.read_csv('../data/train_dataset.csv')

print('all columns:')
data.columns
```

```python
'''
作相关性热力图
'''
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False

get1 = ['信用分', '用户实名制是否通过核实', '用户年龄', '是否大学生客户', '是否黑名单客户', '是否4G不健康客户', '用户网龄（月）']
get2 = ['信用分', '用户最近一次缴费距今时长（月）', '缴费用户最近一次缴费金额（元）', '用户近6个月平均消费值（元）', '用户账单当月总费用（元）', '用户当月账户余额（元）', '缴费用户当前是否欠费缴费']
get3 = ['信用分', '用户话费敏感度', '当月通话交往圈人数', '是否经常逛商场的人', '近三个月月均商场出现次数', '当月是否逛过福州仓山万达', '当月是否到过福州山姆会员店', '当月是否看电影', '当月是否景点游览']
get4 = ['信用分', '当月是否体育场馆消费', '当月网购类应用使用次数', '当月物流快递类应用使用次数', '当月金融理财类应用使用总次数', '当月视频播放类应用使用次数', '当月飞机类应用使用次数', '当月火车类应用使用次数', '当月旅游资讯类应用使用次数']

# plot1
plt.figure(1, figsize=(10,8), dpi= 200)
sns.heatmap(data[get1].corr(), xticklabels=data[get1].corr().columns, yticklabels=data[get1].corr().columns, center=0, annot=True, cmap='GnBu')
plt.title('信用分和其他特征相关图1', fontsize=14)
plt.xticks(fontsize=10, rotation=30)
plt.yticks(fontsize=10, rotation=0)
plt.show()

# plot2
plt.figure(2, figsize=(10,8), dpi= 200)
sns.heatmap(data[get2].corr(), xticklabels=data[get2].corr().columns, yticklabels=data[get2].corr().columns, center=0, annot=True, cmap='GnBu')
plt.title('信用分和其他特征相关图2', fontsize=14)
plt.xticks(fontsize=10, rotation=30)
plt.yticks(fontsize=10, rotation=0)
plt.show()

# plot3
plt.figure(3, figsize=(10,8), dpi= 200)
sns.heatmap(data[get3].corr(), xticklabels=data[get3].corr().columns, yticklabels=data[get3].corr().columns, center=0, annot=True, cmap='GnBu')
plt.title('信用分和其他特征相关图3', fontsize=14)
plt.xticks(fontsize=10, rotation=30)
plt.yticks(fontsize=10, rotation=0)
plt.show()

# plot4
plt.figure(4, figsize=(10,8), dpi= 200)
sns.heatmap(data[get4].corr(), xticklabels=data[get4].corr().columns, yticklabels=data[get4].corr().columns, center=0, annot=True, cmap='GnBu')
plt.title('信用分和其他特征相关图4', fontsize=14)
plt.xticks(fontsize=10, rotation=30)
plt.yticks(fontsize=10, rotation=0)
plt.show()
```

```python
'''
作强弱特征散点图
'''
credit_score = data['信用分']
net_age = data['用户网龄（月）']
ave_cost = data['用户近6个月平均消费值（元）']
express_app_use = data['当月物流快递类应用使用次数']
show_times_in_shop = data['近三个月月均商场出现次数']

#plot1
plt.style.use('ggplot')
plt.figure(5, figsize=(12, 10), dpi=200)
plt.scatter(net_age, credit_score, c='steelblue', s=2)
plt.title('网龄&信用分散点图', fontsize=16)
plt.ylabel('信用分')
plt.xlabel('用户网龄（月）')

#plot2
plt.style.use('ggplot')
plt.figure(6, figsize=(12, 10), dpi=200)
plt.scatter(ave_cost, credit_score, c='steelblue', s=2)
plt.title('消费&信用分散点图', fontsize=16)
plt.ylabel('信用分')
plt.xlabel('用户近6个月平均消费值（元）')

#plot3
plt.style.use('ggplot')
plt.figure(7, figsize=(12, 10), dpi=200)
plt.scatter(express_app_use, credit_score, c='steelblue', s=2)
plt.title('快递应用使用次数&信用分散点图', fontsize=16)
plt.ylabel('信用分')
plt.xlabel('当月物流快递类应用使用次数')
plt.xlim([0, 250])  # 由于数据有的错误从而过于离谱，限制x轴范围，便于观察

#plot4
plt.style.use('ggplot')
plt.figure(8, figsize=(12, 10), dpi=200)
plt.scatter(show_times_in_shop, credit_score, c='steelblue', s=2)
plt.title('商场出现次数&信用分散点图', fontsize=16)
plt.ylabel('信用分')
plt.xlabel('近三个月月均商场出现次数')
```

```python
# 信用分频数分布图
bar_data = data.groupby('信用分', as_index=True)['信用分'].count()
bar_x = pd.Series(bar_data.index)
bar_y = pd.Series(bar_data.values)

bar_data = pd.concat([bar_x, bar_y], axis=1)
bar_data.columns = ['信用分', '频数']
# plot1
plt.figure(9, figsize=(12, 10), dpi=200)
sns.scatterplot(x='信用分', y='频数', data=bar_data, palette='Blues')
plt.title('信用分频数分布图')
bar_data.describe() # 常用统计
```
